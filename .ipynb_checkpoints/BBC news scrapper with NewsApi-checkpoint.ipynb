```python
BBC_all_news_df = pd.DataFrame.from_dict(all_articles['articles'])
BBC_all_news_df['Date/Time'] =  BBC_all_news_df['publishedAt']
BBC_all_news_df['news_text'] = BBC_all_news_df['content']
BBC_all_news_df = BBC_all_news_df.drop(['author', 'description', 'content', 'publishedAt', 'source', 'title', 'url', 'urlToImage'], axis=1)
list(BBC_all_news_df.columns)
```

```python
BBC_all_news_df.info()
```

```python
import re
def clean_text(x):
    pattern = r'[^a-zA-z0-9\s]'
    text = re.sub(pattern, '', str(x))
    return text

def clean_all_text(df):
    for column in df.columns.tolist():
        df[column] = df[column].str.replace('\n','')
        df[column] = df[column].str.replace('\r','')
    for index, item in df.iterrows():
        cleantext = clean_text(item['news_text'])
        item['news_text'] = cleantext
    missing = df.isnull().values.any()
    if missing:
        df.dropna(axis=1)
    return df
```

```python
BBC_all_news_df = clean_all_text(BBC_all_news_df)
```

```python
BBC_all_news_df = BBC_all_news_df[BBC_all_news_df['news_text'] != 'The latest five minute news bulletin from BBC World Service']
BBC_all_news_df['news_text'] = BBC_all_news_df['news_text'].str.replace('Image copyrightGetty', ' ')
BBC_all_news_df['news_text'] = BBC_all_news_df['news_text'].str.replace('ImagesImage caption', ' ')
BBC_all_news_df['news_text'] = BBC_all_news_df['news_text'].str.replace('Image copyright', ' ')
BBC_all_news_df['news_text'] = BBC_all_news_df['news_text'].str.replace('Image caption', ' ')
BBC_all_news_df['news_text'] = BBC_all_news_df['news_text'].str.replace('None', '')
```
